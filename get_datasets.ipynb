```python
import pandas as pd

def remap_columns(df):
    df = df.rename(columns={df.columns[0] : 'text', df.columns[1] : 'classification'})
    return df
```

```python
csv_file = 'data/raw/HateBR.csv'
df = pd.read_csv(csv_file)[['instagram_comments', 'offensive_language']]
df['offensive_language'] = df['offensive_language'].astype(bool)
hatebr_dataset = remap_columns(df)
hatebr_dataset
```

```python
csv_file = 'data/raw/ToLD-BR.csv'
df = pd.read_csv(csv_file)
# Qualquer pontuação nas colunas de ofensividade leva o texto a ser considerado discurso de ódio
hatespeech_cls = (df.iloc[:, 1:] > 0).any(axis=1)
df = pd.concat([df.iloc[:,0], hatespeech_cls], axis=1)
toldbr_dataset = remap_columns(df)
toldbr_dataset
```

```python
hatebr_dataset.to_csv('data/hatebr.csv')
toldbr_dataset.to_csv('data/toldbr.csv')
```
